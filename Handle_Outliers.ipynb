```python
import numpy as np 
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
Q1 = np.percentile(arr,25)  # 25% of arr
Q3 = np.percentile(arr,75)  # 75% of arr

```

```python
IQR = Q3 - Q1   # Interquartile range
IQR
```

```python
UF = Q3 + (1.5*IQR) # Upper Fence
LF = Q1 - (1.5*IQR) # Lower Fence
```

```python
l = []                              # To Remove the Outliers
for i in arr:
    if i >= LF and i <= UF:
        l.append(i)
arr2 = np.array(l)            
```

```python
print(arr)
print(arr2)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.boxplot(x = arr2)
```

```python
# If we not remove the outliers , the model will be biased towards the outliers.

sns.boxplot(x=arr)
```
